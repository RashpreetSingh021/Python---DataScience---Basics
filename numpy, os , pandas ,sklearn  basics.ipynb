```python
# to deal with null by imputing mean/median/mode values on null
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer    #to compute fit and transform the values and apply *mean* on them or *median* or *most frequent*

df = pd.read_csv('datafile.csv')
#df.isnull().sum()                    #used to check and give no of null values in each categories

impute = SimpleImputer(missing_values = np.nan, strategy = 'mean')

impute.fit(df.iloc[ : , 2:3 ])


df.iloc[:,2:3] = impute.transform(df.iloc[ : , 2:3 ])  #to get values of rating locating at col 2nd


df.isnull().sum()

df = df.dropna()  #to drop is null found

df.isnull().sum()

df.head(20)  #standard 5..or given user defiend value
```

```python
#this to deal full data's null value with imputing mode(most frequent) at d place of null

import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer    #to compute fit and transform the values and apply *mean* on them or *median* or *most frequent*

df = pd.read_csv('datafile.csv')
#df
df.isnull().sum()                    #used to check and give no of null values in each categories

impute = SimpleImputer(missing_values = np.nan, strategy = 'most_frequent')

impute.fit(df.iloc[ : , : ])


df.iloc[:,:] = impute.transform(df.iloc[ : , : ])  #to get values of rating locating at col 2nd


df.isnull().sum()
```
